# Dimensionality Reduction

see also:

* https://en.wikipedia.org/wiki/Nonlinear_dimensionality_reduction
* https://scikit-learn.org/stable/modules/manifold.html
* https://scikit-learn.org/stable/auto_examples/manifold/plot_lle_digits.html

```python
from time import time
import numpy as np
import pandas as pd
import sklearn.datasets as datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X, y = datasets.load_breast_cancer(return_X_y=True)
```

```python
def plot2Dtrans(X, y, trans, title=None):
    X_trans = np.transpose(trans.fit_transform(X))
    sns.scatterplot(X_trans[0], X_trans[1], hue=y, alpha=0.5)
    plt.title(title)
```

# PCA

```python
from sklearn.decomposition import PCA

plot2Dtrans(X, y, PCA(n_components=2), 'Principal Component Analysis (PCA)')
```

# Local Linear Embedding

```python
from sklearn.manifold import LocallyLinearEmbedding

plot2Dtrans(X,y,
            LocallyLinearEmbedding(n_components=2, method = 'modified', random_state=42),
            'Local Linear Embedding')
```

# t-SNE

```python
from sklearn.manifold import TSNE

# TODO: needs optimization of parameters (perplexity etc.)
plot2Dtrans(X,y,
            TSNE(n_components=2, random_state=42),
            't-SNE')
```
